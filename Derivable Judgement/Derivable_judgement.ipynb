### 🤔 Q1. Formulate at least two hypotheses from the dataset. Example:
* H₀: 🚬 Smoking has no effect on Diabetes prevalence.
* H₁: 🚬 Smoking affects Diabetes prevalence.


```python
table1 = pd.crosstab(df['smoking_status'], df['diabetes'])
table1
```

```python
chi2, p_value, dof, expected = chi2_contingency(table1)
print("Chi-square value:", chi2)
print("P-value:", p_value)

print("---------------------------------------------------------------")

if p_value < 0.05:
    print("Reject H0: Smoking status affects diabetes prevalence.")
else:
    print("Accept H0: Smoking status has no significant effect on diabetes.")
```

### 🔍 Interpretation:

The p-value (**0.345**) is greater than the significance level (**0.05**), so the difference in diabetes rates across Former Smoker, Non-Smoker, and Smoker groups is **not statistically significant** — it could easily have occurred by chance.

### ✅ Conclusion:

We **accept H₀** — smoking status does not significantly affect diabetes prevalence in this sample. We **cannot conclude** that smoking and diabetes are related here.


#### 🧩 Perform ANOVA Test

```python
daily = df[df['exercise_frequency']=="Daily"]['bmi']
weekly = df[df['exercise_frequency']=="Weekly"]['bmi']
rarely = df[df['exercise_frequency']=="Rarely"]['bmi']
never = df[df['exercise_frequency']=="Never"]['bmi']
```

```python
groups = [daily, weekly, rarely, never]

k = len(groups)        # number of groups

N = sum(len(group) for group in groups)

print("Number of groups:", k)
print("Total observations:", N)
```

```python
all_bmi = pd.concat(groups)

overall_mean = all_bmi.mean()

print("Overall Mean BMI:", overall_mean)
```

```python
SSB = 0

for group in groups:
    group_size = len(group)
    group_mean = group.mean()
    
    SSB += group_size * (group_mean - overall_mean)**2

print("Between Group Sum of Squares:", SSB)

```

```python
SSW = 0

for group in groups:
    group_mean = group.mean()
    
    SSW += sum((group - group_mean)**2)

print("Within Group Sum of Squares:", SSW)
```

```python
df_between = k - 1

df_within = N - k

print("DF Between:", df_between)
print("DF Within:", df_within)
```

```python
MSB = SSB / df_between

MSW = SSW / df_within


print("Mean Square Between:", MSB)

print("Mean Square Within:", MSW)
```

```python
F_value = MSB / MSW

print("F Statistic:", F_value)
```

```python
p_value = 1 - f.cdf(F_value, df_between, df_within)

print("P-value:", p_value)

alpha = 0.05

print("------------------------------------------------------------")

if p_value < alpha:
    print("Reject H0")
    print("Exercise frequency has a significant effect on BMI.")
    
else:
    print("Accept H0")
    print("Exercise frequency does not have a significant effect on BMI.")
```

### 🔍 Interpretation:

The F-statistic (**0.609**) is far below the critical value (**2.623**), and the p-value (**0.610**) is well above **0.05**, meaning exercise frequency explains almost none of the variation in BMI.

### ✅ Conclusion:

We **accept H₀** — exercise frequency (Daily/Weekly/Rarely/Never) does **not** have a significant effect on BMI in this sample.


### 2️⃣ Calculate Confidence Intervals for key numerical data like age, weight, etc.


```python
columns = [
    'age',
    'weight',
    'blood_pressure',
]
```

```python
confidence = 0.95

for column in columns:
    
    data = df[column]
    
    # Number of observations
    n = len(data)
    
    # Mean
    mean = data.mean()
    
    # Standard deviation
    std = data.std()
    
    # Standard error
    standard_error = std / np.sqrt(n)
    
    # t critical value
    t_value = t.ppf((1 + confidence) / 2, n-1)
    
    # Margin of error
    margin_error = t_value * standard_error
    
    # Confidence interval
    lower_limit = mean - margin_error
    upper_limit = mean + margin_error
    
    print("\nColumn:", column)
    print("Mean:", round(mean,2))
    print("95% Confidence Interval:",
          round(lower_limit,2),
          "to",
          round(upper_limit,2))
```

### 🔍 Interpretation:

We are **95% confident** the true population mean lies within each interval below:
- 🎂 **Age:** 41.09 to 44.26 years (mean ≈ 42.67)
- ⚖️ **Weight:** 75.8 to 79.21 kg (mean ≈ 77.51)
- ❤️ **Blood Pressure:** 135.63 to 140.06 (mean ≈ 137.84)

### ✅ Conclusion:

All three intervals are **narrow**, so the sample mean estimates are **precise and reliable**. With n = 500, we have good confidence these ranges reflect the true population values.

### 4️⃣ Perform z-test or t-test based on sample size (mean comparison across groups).


```python
male_bmi = df[df['gender']=="Male"]['bmi']

female_bmi = df[df['gender']=="Female"]['bmi']
```

```python
n1 = len(male_bmi)
n2 = len(female_bmi)

print("Male sample size:", n1)
print("Female sample size:", n2)
```

```python
mean1 = male_bmi.mean()

mean2 = female_bmi.mean()

print("Male BMI Mean:", mean1)

print("Female BMI Mean:", mean2)
```

```python
var1 = male_bmi.var()

var2 = female_bmi.var()
```

```python
standard_error = np.sqrt(
    (var1/n1) + (var2/n2)
)
```

```python
t_value = (mean1 - mean2) / standard_error

print("T Statistic:", t_value)
```

```python
df_value = n1 + n2 - 2

print("Degrees of Freedom:", df_value)
```

```python
p_value = 2 * (1 - t.cdf(abs(t_value), df_value))

print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
    print("There is a significant difference between male and female BMI.")
    
else:
    print("Accept H0")
    print("There is no significant difference between male and female BMI.")
```

### 🔍 Interpretation:

t = **1.287**, df = 333, p-value = **0.199** — well above **0.05**, so the gap between male BMI (26.69) and female BMI (26.02) is **not statistically significant**.

### ✅ Conclusion:

We **accept H₀** — there is **no significant difference** between male and female BMI in this sample.


### 5️⃣ Conduct a chi-square test on categorical data (e.g., 🚬 Smoking habit vs Disease).


```python
table = pd.crosstab(
    df['smoking_status'],
    df['diabetes']
)

table
```

```python
row_total = table.sum(axis=1)

column_total = table.sum(axis=0)

total = table.sum().sum()


expected = np.outer(row_total, column_total) / total


expected = pd.DataFrame(
    expected,
    index=table.index,
    columns=table.columns
)


expected
```

```python
chi_square = ((table - expected)**2 / expected).sum().sum()

print("Chi-square value:", chi_square)
```

```python
rows = table.shape[0]

columns = table.shape[1]


df_value = (rows-1)*(columns-1)


print("Degrees of Freedom:", df_value)
```

```python
p_value = chi2.sf(
    chi_square,
    df_value
)
```

```python
alpha = 0.05
critical_value = chi2.ppf(
    1-alpha,
    df_value
)


print("Critical Value:", critical_value)
```

```python
if p_value < alpha:
    
    print("Reject H0")
    print("Smoking status and diabetes are significantly related.")

else:
    
    print("Accept H0")
    print("Smoking status and diabetes are independent.")
```

### 🔍 Interpretation:

This manual chi-square calculation matches the earlier `chi2_contingency` result (χ² = **2.131**, p = **0.345**), confirming the test is correct.

### ✅ Conclusion:

We **accept H₀** — smoking status and diabetes are **independent**; smoking status does not predict diabetes here.


###  6️⃣ Perform an ANOVA test to check if age groups significantly differ in disease rate.


```python
df['disease_rate'] = df['diabetes'].astype(int)
```

```python
age_groups = [
    df[df['age_group']=="18-25"]['disease_rate'],
    df[df['age_group']=="26-35"]['disease_rate'],
    df[df['age_group']=="36-45"]['disease_rate'],
    df[df['age_group']=="46-60"]['disease_rate'],
    df[df['age_group']=="60+"]['disease_rate']
]
```

```python
k = len(age_groups)

N = sum(len(group) for group in age_groups)


print("Number of groups:", k)

print("Total observations:", N)
```

```python
all_data = pd.concat(age_groups)

overall_mean = all_data.mean()


print("Overall Disease Rate:", overall_mean)
```

```python

SSB = 0

for group in age_groups:
    
    n = len(group)
    
    group_mean = group.mean()
    
    SSB += n * (group_mean - overall_mean)**2


print("SSB:", SSB)
```

```python
SSW = 0


for group in age_groups:
    
    group_mean = group.mean()
    
    SSW += sum((group - group_mean)**2)


print("SSW:", SSW)
```

```python
df_between = k-1

df_within = N-k


print(df_between)

print(df_within)
```

```python
MSB = SSB / df_between

MSW = SSW / df_within


print("MSB:", MSB)

print("MSW:", MSW)
```

```python
F_value = MSB / MSW


print("F-value:", F_value)
```

```python
alpha = 0.05


critical_value = f.ppf(
    1-alpha,
    df_between,
    df_within
)


print("Critical Value:", critical_value)
```

```python
p_value = 1 - f.cdf(
    F_value,
    df_between,
    df_within
)


print("P-value:", p_value)
```

```python
if p_value < alpha:
    
    print("Reject H0")
    print("Disease rate differs significantly among age groups.")

else:
    
    print("Accept H0")
    print("Disease rate does not differ significantly among age groups.")
```

### 🔍 Interpretation:

F = **1.072**, p-value = **0.370** — above **0.05**, so diabetes rate does not vary significantly across the five age groups (18-25 through 60+).

### ✅ Conclusion:

We **accept H₀** — age group does **not** significantly affect diabetes rate in this sample, even though age is normally a known risk factor for diabetes in real-world data.


### 7️⃣ Calculate Covariance and Correlation between continuous variables (e.g., Age vs. BMI)


```python
age = df['age']
bmi = df['bmi']
```

```python
age_mean = age.mean()
bmi_mean = bmi.mean()

print("Mean Age:", age_mean)
print("Mean BMI:", bmi_mean)
```

```python
n = len(df)


covariance = sum(
    (age - age_mean) *
    (bmi - bmi_mean)) / (n-1)


print("Covariance:", covariance)
```

```python
age_std = age.std()

bmi_std = bmi.std()
```

```python
correlation = covariance / (age_std * bmi_std)


print("Correlation:", correlation)
```

### 🔍 Interpretation:

The correlation coefficient (r = **0.032**) is very close to **0**, indicating **almost no linear relationship** between age and BMI.

### ✅ Conclusion:

Age and BMI are **not meaningfully correlated** in this dataset — a person's age gives almost no information about their likely BMI.
